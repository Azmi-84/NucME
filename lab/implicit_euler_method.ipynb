```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, y):
    return ((np.sin(y)-np.exp(t))/np.cos(t))

def df(t, y):
    return (np.cos(y)/np.cos(t))
```

```python
def euler_backward_method(a, b, h, y0):
    number_of_steps = int(round((b-a)/h)+1)

    t_values = np.linspace(a, b, number_of_steps)
    y_values = np.zeros(number_of_steps)
    
    y_values[0] = y0

    for i in range(number_of_steps-1):
        z_guess = y_values[i] 
        
        for _ in range(10):
            g = z_guess - y_values[i] - h * f(t_values[i+1], z_guess)
            dg = 1 - h * df(t_values[i+1], z_guess)
            z_guess = z_guess - g / dg
            
        y_values[i+1] = z_guess
    
    print(f"Number of steps: {number_of_steps}")
    print(f"Time steps: {t_values}")
    print(f"Estimated solutions for each time step: {y_values}")

    plt.figure(figsize=(10, 6))
    plt.plot(t_values, y_values, '.', markersize=10, color='red', 
             label=f'Backward Euler (h = {h})')
    plt.title(f'Backward Euler Method: y(t) vs Time')
    plt.xlabel('Time t')
    plt.ylabel('y(t)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
euler_backward_method(0, 1, 0.5, 0)
```

```python
euler_backward_method(0, 1, 0.02, 0)
```

```python
euler_backward_method(0, 1, 0.001, 0)
```
