```python
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
```

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'k', linewidth=2)
plt.title("Sin Curve")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)

_polynomials_cofficients = [1, 0]
_yt = np.polyval(_polynomials_cofficients, x)
plt.plot(x, _yt, 'b--', linewidth=2)

_polynomials_cofficients = [-1/math.factorial(3), 0, 1, 0]
_yt = np.polyval(_polynomials_cofficients, x)
plt.plot(x, _yt, 'r--', linewidth=2)

_polynomials_cofficients = [1/math.factorial(5), 0, -1/math.factorial(3), 0, 1, 0]
_yt = np.polyval(_polynomials_cofficients, x)
plt.plot(x, _yt, 'g--', linewidth=2)

_polynomials_cofficients = [-1/math.factorial(7), 0, 1/math.factorial(5), 0, -1/math.factorial(3), 0, 1, 0]
_yt = np.polyval(_polynomials_cofficients, x)
plt.plot(x, _yt, 'y--', linewidth=2)

_polynomials_cofficients = [1/math.factorial(9), 0, -1/math.factorial(7), 0, 1/math.factorial(5), 0, -1/math.factorial(3), 0, 1, 0]
_yt = np.polyval(_polynomials_cofficients, x)
plt.plot(x, _yt, 'c--', linewidth=2)

plt.legend(('sin(x)', '1st order', '3rd order', '5th order', '7th order', '9th order'))
plt.show()
```
