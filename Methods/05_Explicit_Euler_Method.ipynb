```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, y):
    return ((np.sin(y)-np.exp(t))/np.cos(t))
```

```python
def euler_forward_method(a, b, h, y0):
    number_of_step = int(round(((b-a)/h)+1))

    t_value = np.linspace(a, b, number_of_step)
    y_value = np.zeros(number_of_step)

    t_value[0] = a
    y_value[0] = y0

    for i in range(1, number_of_step):
        y_value[i] = y_value[i-1] + f(t_value[i-1], y_value[i-1])*h
    
    print(f"Number of steps: {number_of_step}")
    print(f"Time steps: {t_value}")
    print(f"Estimated solutions for each time step: {y_value}")

    plt.figure(figsize=(10, 6))
    plt.plot(t_value, y_value, '.', markersize=10, color='red', label='Forward Euler Method $h=0.5$')
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()
```

```python
euler_forward_method(0, 1, 0.5, 0)
```

```python
euler_forward_method(0, 1, 0.02, 0)
```

```python
euler_forward_method(0, 1, 0.001, 0)
```
